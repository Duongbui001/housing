```python
df= pd.read_excel('Dwellings_total.xlsx', sheet_name='Dwellings', skiprows=[0,2], index_col=0, usecols='B,V', nrows=33)
df.columns = ['House number']
df.index.name = 'Borough'
df
```

```python
ax = df.groupby('Borough').sum().sort_values(by='House number').plot.barh(legend=False, figsize=(10,8))
ax.set_xlabel('Number of houses per borough')
ax.set_title('Distribution of houses in London boroughs in 2020')
plt.show()
```

```python
df= pd.read_excel('Dwellings_total.xlsx', sheet_name='Population', skiprows=[0,2], index_col=0, usecols='B,AR', nrows=33)
df.columns = ['Population']
df.index.name = 'Borough'
df
```

```python
ax = df.groupby('Borough').sum().sort_values(by='Population').plot.barh(legend=False, figsize=(10,8), color='orange')
ax.set_xlabel('Number of population per borough')
ax.set_title('Distribution of populations in London boroughs in 2020')
plt.show()
```
